```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
parsimony = 0.845
nj = 0.03
weighbor = 6.89

dat = pd.DataFrame(data = {"Parsimony" : parsimony, "NJ" : nj, "Weighbor" : weighbor}, index = [0])
dat
```

```python
# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center',)
```

```python
plt.bar(x = dat.columns, height = dat.iloc[0])
plt.title("Running Times for of Different Phylogenetic Algorithms")
plt.xlabel("Phylogenetic Algorithms")
plt.ylabel("Time (seconds)")
addlabels(dat.columns, [parsimony, nj, weighbor])
plt.show()
```
