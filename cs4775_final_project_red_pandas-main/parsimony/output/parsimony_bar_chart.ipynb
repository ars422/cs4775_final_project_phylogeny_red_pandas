```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bayesian = 130
nj = 131
weighbor = 307

dat = pd.DataFrame(data = {"Bayesian" : bayesian, "NJ" : nj, "Weighbor" : weighbor}, index = [0])
dat
```

```python
# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center',)
```

```python
plt.bar(x = dat.columns, height = dat.iloc[0])
plt.title("Parsimony Scores for Trees of Different Phylogenetic Algorithms")
plt.xlabel("Phylogenetic Algorithms")
plt.ylabel("Parsimony Score")
addlabels(dat.columns, [bayesian, nj, weighbor])
plt.show()
```
